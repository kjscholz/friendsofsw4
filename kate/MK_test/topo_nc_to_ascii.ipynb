```python
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
from obspy.geodetics.base import gps2dist_azimuth
```

```python
lat_min=19.2
lat_max=20.0
lon_min=-155.9
lon_max=-155.12
domain_x_bot,_, _=gps2dist_azimuth(lat_min, lon_min, lat_min, lon_max)
domain_y_left,_,_=gps2dist_azimuth(lat_min, lon_min, lat_max, lon_min)
domain_x_top,_, _=gps2dist_azimuth(lat_max, lon_min, lat_max, lon_max)
domain_y_right,_,_=gps2dist_azimuth(lat_min, lon_max, lat_max, lon_max)
print(domain_x_bot)
print(domain_x_top)
print(domain_y_left)
print(domain_y_right)
      
```

# Convert to pandas dataframe

```python
df_grid = grid.to_dataframe().reset_index().sort_values('lat')
df_grid
```

# save to csv

* custom first line with lon/lat to respect formating
* no header of index

```python
custom_first_line = f"{len(grid.lon.data)} {len(grid.lon.data)}"

# Specify the file path where you want to save the CSV file
file_path = 'maunakea.topo'

# Open the file in write mode and add the custom first line
with open(file_path, 'w') as file:
    file.write(custom_first_line + '\n')

# Append the DataFrame to the CSV file (including headers)
df_grid.to_csv(file_path, mode='a', index=False, header=False)
```
